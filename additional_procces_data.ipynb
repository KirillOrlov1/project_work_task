```python
import os
from sklearn.model_selection import train_test_split
import csv
import pandas as pd
```

```python
path = '/content/background_noise_dataset/content/background_noise_dataset'
data = []
data.extend(os.listdir(path + '/' + 'noise'))
data.extend(os.listdir(path + '/' + 'no_noise'))
print('dataset size', len(data))
```

```python
true_preds = []
for item in data:
    if item.startswith('no_'):
        true_preds.append(0)
    else:
        true_preds.append(1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(data, true_preds, test_size=0.3)
```

```python
with open('/content/drive/MyDrive/train.csv', 'w') as csvfile:

    writer = csv.writer(csvfile)
    writer.writerow(['file_name', 'noise'])
    
    for i in range(len(X_train)):
        writer.writerow([X_train[i], y_train[i]])
```

```python
with open('/content/drive/MyDrive/test.csv', 'w') as csvfile:

    writer = csv.writer(csvfile)
    writer.writerow(['file_name', 'noise'])
    
    for i in range(len(X_test)):
        writer.writerow([X_test[i], y_test[i]])
```
